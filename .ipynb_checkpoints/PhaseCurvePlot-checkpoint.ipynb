```python
#planck function
h = 6.626e-34
c = 3.0e+8
k = 1.38e-23
wav = 7.5e-6

def planck(T):
    a = 2.0*h*c**2
    b = h*c/(wav*k*T)
    intensity = a/ ( (wav**5) * (np.exp(b) - 1.0) )
    return intensity
```

```python
pc_telescope.columns
```

```python
def ESM(T_day,T_star, Rp, R_star, mag):
    b_ratio = planck(T_day)/planck(T_star)
    r_ratio = ((7.1492e+7 *Rp)/(6.957e+8*R_star))**2
    mag_ratio = 10**(-mag/5) 
    esm = 4.29e6*b_ratio*r_ratio*mag_ratio
    return esm
```

```python
pc_telescope['ESM'] = ESM(1.1*pc_telescope['pl_eqt'], pc_telescope['st_teff'],pc_telescope['pl_radj'] ,pc_telescope['st_rad'],pc_telescope['sy_kmag'])
```

```python
# g = GM/r^2
G = 6.67e-11
M_jup = 1.89e27
r_jup = 69911*1000

pc_telescope['pl_g'] = (G*M_jup*pc_telescope['pl_bmassj'])/ ((r_jup*pc_telescope['pl_radj'])**2)
print(G*M_jup/ (r_jup**2))
```
